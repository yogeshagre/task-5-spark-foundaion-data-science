```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ipl = pd.read_csv(r'C:\Users\ysa25\Desktop\matches.csv')

ipl.head()
```

```python
#determine the no.of rows and columns in dataset
ipl.shape
```

```python
#finding the frequency of most man of the match in the match
ipl['player_of_match'].value_counts()
```

```python
ipl['player_of_match'].value_counts()[0:10]
```

```python
ipl['player_of_match'].value_counts()[0:5]
```

```python
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
#making bar plot for most man of the match in ipl
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color='g')
plt.show()
```

```python
#getting frequency of result column
ipl['result'].value_counts()
```

```python
#finding no.toss wins by a team in ipl 
ipl['toss_winner'].value_counts()
```

```python
#records when team won the batting first
batting_first=ipl[ipl['win_by_runs']!=0]
batting_first

```

```python
#making a histogram
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title('distribution of runs')
plt.xlabel('runs')
plt.ylabel('matches')
plt.show()
```

```python
#finding no.of wins w.r.t.each team batting first
batting_first['winner'].value_counts()
```

```python
#making a bar plot for top 3 team after win by batting first
plt.figure(figsize=(5,5))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),(batting_first['winner'].value_counts()[0:3]),color=['blue','yellow','red'])
plt.show()
```

```python
#making a pie chart
plt.figure(figsize=(5,5))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()           
           
```

```python
#finding those records when batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
batting_second
```

```python
#at the head
batting_second.head()
```

```python
#making a histogram for frequency of wins number of wickets
plt.figure(figsize=(5,5))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
#finding out frequency of wins w.r.t.after batting second
batting_second['winner'].value_counts()
```

```python
#making a bar plot for top 3 teams after winning by  taking batting second
plt.figure(figsize=(5,5))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=['purple','blue','red'])
plt.show()

```

```python
#making a pie chart for distribution of most wins by batting second
plt.figure(figsize=(5,5))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
